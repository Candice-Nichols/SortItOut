```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#differences between pdb and em map (new equation)
import seaborn as sns
import pandas as pd
diff_p = pd.read_csv("diff_plot4.csv")
sns.set(style="whitegrid",rc={'figure.figsize':(11,8)},font_scale=1.5)
sns.set_palette("Set3")
ax = sns.violinplot(x=diff_p["group"],y=diff_p["diff pdb em"],hue=diff_p["Linreg"],split=True,inner="quartile") 
ax.set_title("Differences Between Atomic Volume and EM Volume",fontsize=25)
ax.title.set_position([.5, 1.05])
ax.set_xlabel("Group")
ax.set_ylabel("Atomic volume - EM volume")
ax.ticklabel_format(axis="y", style="sci", scilimits=(6,6))
ax.legend(fontsize=18,title="Linear Regression",title_fontsize=18)
fig = ax.get_figure()
#fig.savefig("DiffPlot.jpg")
```

```python
##### pdb volume spread before and after linear regression (new equation)
sns.set(style="whitegrid")
pdb_p = pd.read_csv("pdbvol_linreg4.csv")
sns.set(style="whitegrid",rc={'figure.figsize':(11,8)},font_scale=1.5)
sns.set_palette("Set3")
ax = sns.violinplot(x=pdb_p["GROUP"],y=pdb_p["PDBVOL"],hue=pdb_p["LINREG"],split=True,inner="quartile") 
ax.set_title("Atomic Volume Spread Before And After Linear Regression",fontsize=25)
ax.title.set_position([.5, 1.10])
ax.set_xlabel("Group")
ax.set_ylabel("Atomic Volume")
ax.legend(fontsize=18,title="Linear Regression",title_fontsize=18)
fig = ax.get_figure()
fig.savefig("SpreadPlot.jpg")
```
